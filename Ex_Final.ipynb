```python
# Para calcular métricas de una matriz de confusión para detección de objetos (como rostros),
# necesitamos 'ground truth' (ubicaciones reales de los rostros) en un conjunto de imágenes.
# El código actual detecta rostros en una sola imagen y no tiene ground truth.

# --- Función para calcular Intersection over Union (IoU) ---
# IoU es fundamental para determinar si una detección coincide con un rostro real.
def calculate_iou(boxA, boxB):
    # boxA y boxB son [x, y, w, h]
    # Calcular las coordenadas del rectángulo de intersección
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[0] + boxA[2], boxB[0] + boxB[2])
    yB = min(boxA[1] + boxA[3], boxB[1] + boxB[3])

    # Calcular el área de intersección
    interArea = max(0, xB - xA) * max(0, yB - yA)

    # Calcular el área de ambos bounding boxes
    boxAArea = boxA[2] * boxA[3]
    boxBArea = boxB[2] * boxB[3]

    # Calcular IoU
    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou

# --- Definición de umbral IoU ---
# Un umbral común para considerar que una detección es una 'coincidencia'
iou_threshold = 0.5

# --- Detecciones de la imagen actual ---
# 'faces' proviene de la detección en la celda anterior.
# Asumimos que 'faces' contiene las detecciones del clasificador.
detecciones = faces

# --- Ground Truth para la imagen actual (EJEMPLO CONCEPTUAL) ---
# Para esta demostración, asumiremos un ground truth para 'foto viviana.jpeg'.
# En un escenario real, esta lista provendría de anotaciones manuales.
# Basado en los resultados de 'faces' de la celda anterior, donde el índice 0 era
# probablemente un verdadero positivo y el índice 1 un falso positivo.
ground_truth_boxes = [
    [241, 299, 493, 493] # Asumiendo que esta es la única cara real en la imagen.
]

# --- Inicializar contadores para Verdaderos Positivos (TP), Falsos Positivos (FP), Falsos Negativos (FN) ---
current_tp = 0
current_fp = 0
current_fn = 0

# Para rastrear si un ground truth box ya ha sido 'matcheado' por una detección
matched_gt = [False] * len(ground_truth_boxes)

# --- Comparar detecciones con Ground Truth ---
for det_box in detecciones:
    best_iou = 0
    best_match_idx = -1

    for gt_idx, gt_box in enumerate(ground_truth_boxes):
        iou = calculate_iou(det_box, gt_box)
        if iou > best_iou:
            best_iou = iou
            best_match_idx = gt_idx

    if best_iou >= iou_threshold: # Si la mejor IoU excede el umbral
        if not matched_gt[best_match_idx]: # Y el ground truth box no ha sido matcheado aún
            current_tp += 1
            matched_gt[best_match_idx] = True # Marcar como matcheado
        else:
            # Si un ground truth box es detectado por múltiples detecciones,
            # solo la primera cuenta como TP, las subsiguientes son FP.
            current_fp += 1
    else:
        current_fp += 1 # La detección no cumple el umbral o no tiene coincidencia

# Contar Falsos Negativos (Ground Truths no detectados)
for matched in matched_gt:
    if not matched:
        current_fn += 1

# --- Imprimir métricas para la imagen actual ---
print(f"--- Métricas para la imagen actual (Conceptuales) ---")
print(f"Verdaderos Positivos (TP): {current_tp}")
print(f"Falsos Positivos (FP): {current_fp}")
print(f"Falsos Negativos (FN): {current_fn}")

# --- Calcular Precision, Recall y F1-Score ---
# Precision: TP / (TP + FP) - ¿Cuántas de las detecciones fueron correctas?
# Recall: TP / (TP + FN) - ¿Cuántos de los rostros reales fueron detectados?
# F1-Score: 2 * (Precision * Recall) / (Precision + Recall) - Media armónica de Precision y Recall

precision = current_tp / (current_tp + current_fp) if (current_tp + current_fp) > 0 else 0.0
recall = current_tp / (current_tp + current_fn) if (current_tp + current_fn) > 0 else 0.0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1_score:.2f}")
```

```python
import pandas as pd
import os
import cv2

# --- Definir rutas de tu dataset en Google Drive ---
# Asegúrate que esta es la ruta correcta al *directorio* que contiene ground_truth.csv
dataset_base_path = '/content/drive/MyDrive/codigos/intro_a_python/examen final/mi_dataset_rostros/'
ground_truth_csv_path = os.path.join(dataset_base_path, 'ground_truth.csv')

# --- Cargar el archivo CSV de ground truth ---
try:
    df_ground_truth = pd.read_csv(ground_truth_csv_path)
    print(f"CSV de ground truth cargado exitosamente desde: {ground_truth_csv_path}")
    print("Primeras 5 filas del CSV:")
    print(df_ground_truth.head())
except FileNotFoundError:
    print(f"Error: El archivo ground_truth.csv no se encontró en {ground_truth_csv_path}")
    print("Por favor, asegúrate de que la ruta y el nombre del archivo son correctos y que has creado el dataset como se indicó.")
    df_ground_truth = None

# Función para calcular y devolver métricas
def calculate_metrics(tp, fp, fn):
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
    return precision, recall, f1_score

# Función para imprimir métricas (puede llamarse después de calcular)
def print_metrics(tp, fp, fn, precision, recall, f1_score, label, tn=None):
    print(f"\n--- {label} ---")
    print(f"  Verdaderos Positivos (TP): {tp}")
    print(f"  Falsos Positivos (FP): {fp}")
    print(f"  Falsos Negativos (FN): {fn}")
    if tn is not None:
        print(f"  Verdaderos Negativos (TN): {tn}") # Añadir conteo de TN
    print(f"  Precision: {precision:.2f}")
    print(f"  Recall: {recall:.2f}")
    print(f"  F1-Score: {f1_score:.2f}")


# Solo proceder si df_ground_truth se cargó correctamente
if df_ground_truth is not None:
    # --- Agrupar las anotaciones por imagen para fácil acceso ---
    image_ground_truths = {}
    for index, row in df_ground_truth.iterrows():
        image_name = row['Image Name']
        condition = row['Condition']
        true_positives_count = row['True Positives'] # Obtener el conteo de verdaderos positivos

        image_ground_truths[image_name] = {
            'condition': condition,
            'gt_true_positives': true_positives_count # Almacenar el conteo
        }

    print(f"\nSe han procesado {len(image_ground_truths)} imágenes únicas en el CSV.")
    print("Ejemplo de ground truth para la primera imagen:")
    if image_ground_truths:
        first_image_name = list(image_ground_truths.keys())[0]
        print(f"Imagen: {first_image_name}, Condición: {image_ground_truths[first_image_name]['condition']}, Ground Truth True Positives: {image_ground_truths[first_image_name]['gt_true_positives']}")

    print("\nIniciando procesamiento de imágenes del dataset para agregación de métricas...")

    # Inicializar acumuladores para métricas por condición
    metrics_by_condition = {
        'Natural': {'tp': 0, 'fp': 0, 'fn': 0, 'tn': 0}, # Añadir 'tn'
        'Artificial': {'tp': 0, 'fp': 0, 'fn': 0, 'tn': 0} # Añadir 'tn'
    }

    # Lista para almacenar las métricas de cada imagen
    all_image_metrics = []

    for image_name, data in image_ground_truths.items():
        image_condition = data['condition']
        ground_truth_count = data['gt_true_positives'] # Obtener el conteo de ground truth

        # Construir la ruta completa de la imagen (asumiendo que las imágenes están directamente en dataset_base_path)
        image_path = os.path.join(dataset_base_path, image_name)

        # --- Cargar la imagen ---
        image = cv2.imread(image_path)
        if image is None:
            print(f"Advertencia: No se pudo cargar la imagen {image_path}. Saltando...")
            continue

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # --- Realizar la detección de rostros con faceClassif (ya cargado) ---
        # Asegúrate de que 'faceClassif' esté disponible globalmente o pásalo como argumento.
        detections = faceClassif.detectMultiScale(gray,
                                               scaleFactor=1.1,
                                               minNeighbors=5,
                                               minSize=(30,30))

        detected_count = len(detections)

        # --- Calcular TP, FP, FN, TN para la imagen actual (basado en conteos) ---
        current_tp = 0
        current_fp = 0
        current_fn = 0
        current_tn = 0 # Inicializar current_tn por imagen

        if ground_truth_count > 0:
            current_tp = min(ground_truth_count, detected_count)
            current_fp = max(0, detected_count - ground_truth_count)
            current_fn = max(0, ground_truth_count - detected_count)
        else: # ground_truth_count == 0 (no hay rostros reales en la imagen)
            if detected_count == 0:
                current_tn = 1 # El modelo predijo correctamente que no hay rostros
            else:
                current_fp = detected_count # El modelo detectó rostros donde no los hay

        # Calcular métricas para la imagen actual
        precision, recall, f1_score = calculate_metrics(current_tp, current_fp, current_fn)

        # Almacenar métricas por imagen
        all_image_metrics.append({
            'Image Name': image_name,
            'Condition': image_condition,
            'TP': current_tp,
            'FP': current_fp,
            'FN': current_fn,
            'TN': current_tn, # Añadir TN a las métricas por imagen
            'Precision': precision,
            'Recall': recall,
            'F1-Score': f1_score
        })

        # Acumular métricas para la condición respectiva
        if image_condition in metrics_by_condition:
            metrics_by_condition[image_condition]['tp'] += current_tp
            metrics_by_condition[image_condition]['fp'] += current_fp
            metrics_by_condition[image_condition]['fn'] += current_fn
            metrics_by_condition[image_condition]['tn'] += current_tn # Acumular TN

    # Crear DataFrame a partir de all_image_metrics y guardar en CSV
    df_image_metrics = pd.DataFrame(all_image_metrics)
    output_csv_path = os.path.join(dataset_base_path, 'per_image_metrics.csv')
    df_image_metrics.to_csv(output_csv_path, index=False)
    print(f"\nCSV de métricas por imagen guardado en: {output_csv_path}")
    print("Primeras 5 filas del CSV de métricas por imagen:")
    print(df_image_metrics.head())

    print("\n--- Resultados Agregados por Condición ---")
    for condition, metrics in metrics_by_condition.items():
        agg_precision, agg_recall, agg_f1_score = calculate_metrics(metrics['tp'], metrics['fp'], metrics['fn'])
        print_metrics(metrics['tp'], metrics['fp'], metrics['fn'], agg_precision, agg_recall, agg_f1_score, f"Métricas Agregadas para Condición: {condition}", tn=metrics['tn']) # Pasar TN a la función de impresión

```

### Experimentando con `minSize` para el Clasificador de Rostros

El parámetro `minSize` en `detectMultiScale` es crucial, ya que define el tamaño mínimo (ancho, alto) de los objetos que se detectarán. Un valor incorrecto puede llevar a muchos Falsos Positivos o Falsos Negativos. Realizaremos un barrido de valores para `minSize` y observaremos cómo afecta a las métricas del modelo.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Rango de valores de minSize para probar
# Prueba con un rango significativo, por ejemplo, de 10x10 a 100x100 en pasos de 10
min_sizes_to_test = [(s, s) for s in range(10, 101, 10)]

results_min_size = []

print("\n--- Iniciando barrido de minSize ---")

for min_w, min_h in min_sizes_to_test:
    current_min_size = (min_w, min_h)
    print(f"Evaluando con minSize: {current_min_size}")

    # Reiniciar acumuladores para cada minSize
    metrics_by_condition_current = {
        'Natural': {'tp': 0, 'fp': 0, 'fn': 0, 'tn': 0},
        'Artificial': {'tp': 0, 'fp': 0, 'fn': 0, 'tn': 0}
    }
    total_tp = 0
    total_fp = 0
    total_fn = 0
    total_tn = 0

    for image_name, data in image_ground_truths.items():
        image_condition = data['condition']
        ground_truth_count = data['gt_true_positives']

        image_path = os.path.join(dataset_base_path, image_name)
        image = cv2.imread(image_path)
        if image is None:
            continue # Ya se advirtió antes si no se cargó
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Realizar detección con el minSize actual
        detections = faceClassif.detectMultiScale(gray,
                                               scaleFactor=1.1,
                                               minNeighbors=5,
                                               minSize=current_min_size)
        detected_count = len(detections)

        # Calcular TP, FP, FN, TN para la imagen actual (basado en conteos)
        current_tp_img = 0
        current_fp_img = 0
        current_fn_img = 0
        current_tn_img = 0

        if ground_truth_count > 0:
            current_tp_img = min(ground_truth_count, detected_count)
            current_fp_img = max(0, detected_count - ground_truth_count)
            current_fn_img = max(0, ground_truth_count - detected_count)
        else: # ground_truth_count == 0
            if detected_count == 0:
                current_tn_img = 1
            else:
                current_fp_img = detected_count

        # Acumular para los totales generales
        total_tp += current_tp_img
        total_fp += current_fp_img
        total_fn += current_fn_img
        total_tn += current_tn_img

    # Calcular métricas agregadas para este minSize
    precision, recall, f1_score = calculate_metrics(total_tp, total_fp, total_fn)

    results_min_size.append({
        'minSize': current_min_size[0],
        'TP': total_tp,
        'FP': total_fp,
        'FN': total_fn,
        'TN': total_tn,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1_score
    })

# Convertir resultados a DataFrame para fácil análisis y visualización
df_min_size_results = pd.DataFrame(results_min_size)

print("\n--- Resumen de resultados por minSize ---")
display(df_min_size_results)

# --- Visualización de resultados ---
plt.figure(figsize=(14, 8))

# Plot de TP, FP, FN, TN
plt.subplot(2, 1, 1)
sns.lineplot(x='minSize', y='TP', data=df_min_size_results, label='TP (Verdaderos Positivos)', marker='o')
sns.lineplot(x='minSize', y='FP', data=df_min_size_results, label='FP (Falsos Positivos)', marker='o')
sns.lineplot(x='minSize', y='FN', data=df_min_size_results, label='FN (Falsos Negativos)', marker='o')
sns.lineplot(x='minSize', y='TN', data=df_min_size_results, label='TN (Verdaderos Negativos)', marker='o')
plt.title('Métricas de Detección por Tamaño Mínimo (minSize)')
plt.xlabel('minSize (pixels)')
plt.ylabel('Conteo')
plt.grid(True)
plt.legend()

# Plot de Precision, Recall, F1-Score
plt.subplot(2, 1, 2)
sns.lineplot(x='minSize', y='Precision', data=df_min_size_results, label='Precision', marker='o')
sns.lineplot(x='minSize', y='Recall', data=df_min_size_results, label='Recall', marker='o')
sns.lineplot(x='minSize', y='F1-Score', data=df_min_size_results, label='F1-Score', marker='o')
plt.title('Precision, Recall y F1-Score por Tamaño Mínimo (minSize)')
plt.xlabel('minSize (pixels)')
plt.ylabel('Valor de Métrica')
plt.ylim(0, 1) # Las métricas están entre 0 y 1
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
